# Load .csv file

```python
df = pd.read_csv("/content/cloths-rating (2).csv")
df.head()
```

# Find Sentiment on Text(Reviews)

```python
def sentiment_calc(text):
    try:
        return TextBlob(str(text)).sentiment.polarity
    except:
        return None
df['sentiment'] = df['Text'].apply(sentiment_calc)
df
```

# Apply Multiplication B/W Rating&Sentiment

```python
df['Updated_score'] = df['Rating']*df['sentiment']
df
```

# Make UserId into Normal Form

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['UserID'] = le.fit_transform(df['UserID'])
df
```

# Make function for classify updated_score

```python
classify = {'1':[-5,-4,-3.75,-3.25,3,-2.75,-2.50,-2.25,-2,-1.5,-1.25,-1,-0.75,-0.5,-0.25],
            '2':[-0.24,0.25,0.5,0.75,1] ,
            '3': [1.01,1.25,1.5,1.75,2],
            '4': [2.01,2.25,2.5,2.75,3],
            '5':[3.01,3.25,3.50,3.75,4,5]}

def fun(values):
    for i in classify:
        for j in classify[i]:
            if values == j:
                return i
            if values <= j:
                return i
```

# Apply function on updated_score and put into New_score column

```python
df['New_score'] = df['Updated_score'].apply(fun)
df['New_score'] = pd.to_numeric(df['New_score'])
df
```

# Pivot table of ProductID, UserID and New_score

```python
df_pivot = df.pivot_table(index='ProductID',columns='UserID',values='New_score').fillna(0)
df_pivot
```

# Sparse Matrix (Compressed Sparse Row)

```python
from scipy.sparse import csr_matrix

df_pivot_matrix = csr_matrix(df_pivot.values) 
print(df_pivot_matrix)
```

# Fitting data into NearestNeighborsModel

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', n_neighbors=20, radius=1)
model_knn.fit(df_pivot_matrix)
```

# Matrix of cosine similarity

```python
similarity_matrix = cosine_similarity(df_pivot)
similarity_matrix
```

```python
product_ID = int(input('Enter Product ID according to data set : '))
data = list(df_pivot.index) #shows list of ProductID in data-set
print(data)
```

```python
query_index = data.index(product_ID) #shows index of productID by USER
print(query_index)
```

```python
similarity, indices = model_knn.kneighbors(df_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 8)
print(similarity) #shows similarity distance through productID by USER
print(indices) #shows indexs of productID by USER
```

# Create dictionary and store recommendations of productID which is given by USER

```python
data_dict={}
for i in range(0, len(similarity.flatten())):    #gives length of similarity array

    if i == 0:
        print('Recommendations for {0}:\n'.format(df_pivot.index[query_index]))
    else:
        data_dict[str(df_pivot.index[indices.flatten()[i]])] = float(similarity.flatten()[i])
        print(f'{df_pivot.index[indices.flatten()[i]]}, is similarity distance = with {similarity.flatten()[i]:.20f}:')

print(data_dict)
```

```python
import surprise
from surprise import SVDpp
from surprise import Dataset
from surprise.model_selection import cross_validate
from surprise.model_selection import GridSearchCV
from surprise import Reader
from sklearn.model_selection import  train_test_split
```

```python
df = pd.read_csv("cloths-rating (2).csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['UserID'] = le.fit_transform(df['UserID'])
df
```

```python
df = df.astype(str)
```

```python
df.dtypes
```

# **Fitting the Model (SVDpp)**

```python
reader = surprise.Reader(rating_scale=(1,5))
```

```python
data = Dataset.load_from_df(df[['UserID','ProductID','Rating']],reader)
```

```python
alg = surprise.SVDpp()
```

```python
output = alg.fit(data.build_full_trainset())
```

# **Prediction On model**

```python
pred = alg.predict(uid='3', iid='180')
```

```python
score = pred.est
print(score)
```

# **Making Recommendations**

```python
ids = df['ProductID'].unique()
ids13 = df.loc[df['UserID'] == '13', 'ProductID']
ids_to_predict = np.setdiff1d(ids,ids13)
```

```python
testset = [[13,ProductID,5] for ProductID in ids_to_predict]
predictions = alg.test(testset)
predictions[0]
```

```python
pred_ratings = np.array([pred.est for pred in predictions])
i_max = pred_ratings.argmax()
ProductID = ids_to_predict[i_max]
print("Top Item For User 13 has ProductID {0} With Predicting Rating {1}".format(ProductID,pred_ratings[i_max]))
```

# **Tuning and Evaluating the Model**

```python
param_grid = { 'lr_all': [.001, .01], 'reg_all' : [.1,.5]} 
gs = surprise.model_selection.GridSearchCV(surprise.SVDpp, param_grid, measures=['RMSE', 'MAE'], cv = 5)
gs.fit(data)
print(gs.best_params['rmse'])
```

```python
alg = surprise.SVDpp(lr_all = .001)
output = surprise.model_selection.cross_validate(alg, data, verbose = True)
```
